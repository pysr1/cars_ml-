## Additional EDA

There's multiple cars that are the exact same model but have different MSRPs.

```
df[df['Popularity'] > 5000]['Make'].unique()
```

```
plt.figure(figsize = (12, 7.3))
df.groupby('Make').median()['Popularity'].sort_values().plot(kind = 'barh')
plt.title('Median Popuarity by Make')
plt.xlabel('Median Popularity')
plt.show()
```

```
plt.figure(figsize = (14, 7.1))
sns.boxplot(x = 'Year', y = 'highway MPG', data = df)
```

```
sns.distplot(df['highway MPG'])
```
